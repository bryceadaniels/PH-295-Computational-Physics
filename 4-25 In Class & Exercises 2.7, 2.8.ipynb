### Taylor Series Evaluation
#### Introduction
I did the work by hand to do the series expansion of sine and cosine. 


```python
import numpy as np

cosExp = lambda x: 1-1/2*x**2+1/24*x**4-1/np.math.factorial(6)*x**6+1/np.math.factorial(8)*x**8-1/np.math.factorial(10)*x**10
sinExp = lambda x: x-1/6*x**3+1/120*x**5-1/np.math.factorial(7)*x**7+1/np.math.factorial(9)*x**9-1/np.math.factorial(11)*x**11
print(cosExp(3*np.pi/4),np.cos(3*np.pi/4),(cosExp(3*np.pi/4)-np.cos(3*np.pi/4))/np.cos(3*np.pi/4)*100,'%')
print(sinExp(3*np.pi/4),np.sin(3*np.pi/4),(sinExp(3*np.pi/4)-np.sin(3*np.pi/4))/np.sin(3*np.pi/4)*100,'%')
```

### Conclusion
Depending on how precise we need the calculation to be, we could only include so many terms.

I need to make a program to find the Catalan Numbers up until 1 billion

```python
import numpy as np
C = 1
n = 0
while C <= 1e9:
    print(C)
    C = (4*n+2)/(n+2)*C
    n +=1 
```
